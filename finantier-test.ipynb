```python
import numpy as np 
import pandas as pd 
from sklearn import datasets, neighbors
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('finantier_data_technical_test_dataset.csv', index_col=False)
df
```

```python
# Dropping columns that are not relevant
df.drop('customerID',axis=1,inplace=True)
```

```python
df.info()
```

```python
# Drop all missing values
df.dropna(inplace=True)
df.isna().sum()
```

```python
df
```

```python
# Check for all unique values for each column
def unique_values(tel_df):
    for col in tel_df:
        if tel_df[col].dtypes == 'object':
            print(f'{col} : {tel_df[col].unique()}')
            
unique_values(df)
```

```python
# Replace values
df.replace('No internet service','No',inplace=True)
df.replace('No phone service','No',inplace=True)
unique_values(df)
```

```python
# Using ordinal encoder for transformation of categorical variables
ordinalencoder = OrdinalEncoder()
oe_cols = ['gender','Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup',
                  'DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Default']
for col in oe_cols:
    df[col] = ordinalencoder.fit_transform(df[[col]])
```

```python
df = pd.get_dummies(data=df, columns=['InternetService','Contract','PaymentMethod'])
df
```

```python
# Convert total charges from string to numeric values
df = df[df.TotalCharges != ' ']
df.TotalCharges = pd.to_numeric(df.TotalCharges)
```

```python
#Scale to values between 0 and 1 
cols_to_scale = ['tenure','MonthlyCharges','TotalCharges']

scaler = MinMaxScaler()
df[cols_to_scale] = scaler.fit_transform(df[cols_to_scale])
```

```python
# SPlit the data into train and test sets
X = df.drop(['Default'],axis=1)

y = df.Default
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
# Using Logistic Regression model for prediction
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
result = model.fit(X_train, y_train)
from sklearn import metrics
prediction_test = model.predict(X_test)
# Print the prediction accuracy
print (metrics.accuracy_score(y_test, prediction_test))
```

```python
# Prediction using SVM
from sklearn.svm import SVC

model.svm = SVC(kernel='linear') 
model.svm.fit(X_train,y_train)
preds = model.svm.predict(X_test)
metrics.accuracy_score(y_test, preds)
```

```python
# Evaluation metrics
from sklearn.metrics import classification_report

print(classification_report(y_test, preds))
```

```python
#Confusion Matrix
dlac_conf_matrix = metrics.confusion_matrix(y_test, preds)
print (dlac_conf_matrix)
```
